# 1. EDA용 데이터 구축 : 2019년 실적 데이터 + 시청률 데이터

```python
#데이터 적재
실적_2019 = pd.read_excel('../data/internal/preprocessed_data/train_data/실적_전처리.xlsx') 
rating = pd.read_excel('../data/external/tv_rating/2019_시청률 데이터.xlsx', skiprows=1, skipfooter=1) # 시청률 데이터, 첫째 행 제외, 마지막 행 제외
시청률 = rating.copy()
```

```python
실적_2019.columns
```

```python
실적_2019['판매량'] = (실적_2019.취급액/실적_2019.판매단가).apply(int)
```

```python
#시청률 데이터 마지막 열 없애주기 (마지막 열 = 1년 요약)
시청률 = rating.iloc[:, :-1]
시청률 = 시청률.melt(id_vars=["시간대"], 
        var_name="날짜", 
        value_name="시청률")
시청률['방송일시'] = pd.to_datetime(시청률['날짜'] + ' ' + 시청률['시간대'])

# 시청률 데이터는 2019-01-01 02:00:00부터 시작하지만
# 실적 데이터는 2019-01-01 06:00:00부터 시작하기 때문에
# 시청률 데이터에서 2019-01-01 02:00:00 ~ 2019-01-01 06:00:00의 데이터를 없애줘야 함
from_ts = '2019-01-01 02:00:00'
to_ts = '2019-01-01 06:00:00'
시청률 = 시청률[(시청률['방송일시'] < from_ts) | (시청률['방송일시'] >= to_ts)]

실적_2019['시청률'] = pd.merge_asof(시청률.sort_values('방송일시'), 실적_2019.reset_index(),
                               on='방송일시',
                               direction='backward'
                              ).groupby('index')['시청률'].mean()
```

```python
시청률 = list(실적_2019['시청률'])

for idx, i in zip(실적_2019.index, 시청률):
  if math.isnan(i) == True:
    try:
      시청률[idx] = 시청률[idx+1]
    except:
      print(idx) #idx = 37371
      import statistics as st
      시청률[idx] = st.mean([0,0,0,0,0.024,0.024,0.024,0.024,0.038,0.038,0.024,0.024,0.024,0.024,0.024,0.024,0.024,0.024,0.024,0.024,0.024])
      continue
    

실적_2019['시청률'] = 시청률
```

# 2. EDA 시작!

## A) 상품군별 판매량 Best Items 특징

### - 상품별 베스트 아이템

```python
for i in np.unique(실적.상품군):
    print(i)
    d = 실적[실적.상품군 == i].sort_values('판매량',ascending=False).head(10)[['마더코드','상품코드','취급액','상품명','판매량']]
    print(tabulate(d, headers='keys', tablefmt='psql'))
```

### - 상품군별 베스트 아이템 마더코드

```python
code = dict()
for i in np.unique(실적_2019.상품군):
    print(i)
    d = 실적_2019[실적_2019.상품군 == i].pivot_table(index='마더코드',values=['취급액','판매량'],aggfunc='sum').sort_values('판매량',ascending=False).head(10)
    print(tabulate(d, headers='keys', tablefmt='psql'))
    code[i] = list(d.index)
```

### - Best 10 마더코드에 속하는 상품 수와 총 상품수

```python
for k in code.keys():
    print(k,':',len(실적_2019[실적_2019.마더코드.isin(code[k])]),'(',len(실적_2019[실적_2019.상품군 == k]),')')
```

## B) 시간대 별 취급액 및 시청률  

### - 상품군별 '시'별로 보기

```python
상품군열 = list(실적_2019['상품군'].unique())
상품군열 

item_sum = 실적_2019.groupby(['상품군', '시'])['취급액'].mean()
item_sum=item_sum.to_frame().reset_index()
data = []
for i in 상품군열 :
    s = item_sum.loc[item_sum['상품군']==i]
    trace = go.Scatter(
        x= s.시,
        y= s.취급액,
        name = "상품군 "+str(i),
        opacity = 0.9)
    data.append(trace)

# Buttons to select a specific store visualization
update_buttons = []
for i in 상품군열 :
    visible = [True if j == i else False for j in 상품군열 ]
    button= dict(label = '상품군 ' + str(i), method= 'update', args= [{'visible': visible}])
    update_buttons.append(button)
# Button to return to all stores visualization
update_buttons.append(dict(label = 'All', method= 'update', args= [{'visible': [True]*10}]))

updatemenus = list([dict(active=-1, buttons=list(update_buttons))])

layout = dict(
    title='상품군 시간 별 매출 ',
    updatemenus= updatemenus,
    xaxis=dict(
        # rangeselector=dict(
        #     buttons=list([
        #         dict(count=1, label='1h', step='hour', stepmode='backward'),
        #         dict(count=6, label='6h', step='hour', stepmode='backward'),
        #         dict(step='all')
        #     ])
        # ),
        rangeslider=dict(), type='time'
    )
)

fig = dict(data=data, layout=layout)
iplot(fig, validate= False)
```

### - 월별,상품군별 송출횟수, 노출시간, 판매량 확인

```python
송출횟수 = 실적_2019.pivot_table(index=['상품군'],columns='월',values='상품명',aggfunc='count').T
노출 = 실적_2019.pivot_table(index=['상품군'],columns='월',values='노출분',aggfunc='sum').T
월별판매량 = 실적_2019.pivot_table(index=['상품군'],columns='월',values ='판매량',aggfunc='mean').T
월별취급액 = 실적_2019.pivot_table(index=['상품군'],columns='월',values ='취급액',aggfunc='mean').T

c = 월별판매량.columns
fig, axes = plt.subplots(11,3,figsize=(18,66))
idx=0
color = sns.color_palette("hls", 11)
for i in axes:
    col = color.pop()
    송출횟수[c[idx]].plot(kind='bar',ax=i[0],color=col)
    월별판매량[c[idx]].plot(kind='bar',ax=i[2],color=col)
    노출[c[idx]].plot(kind='bar',ax=i[1],color=col)
    i[0].set(title=c[idx],ylabel='송출횟수')
    i[2].set(title=c[idx],ylabel='판매량')
    i[1].set(title=c[idx],ylabel='노출시간')
    idx+=1
plt.show()
```

```python
cate = 실적_2019.pivot_table(index=['요일'],columns=['상품군'],values = ['판매량'], aggfunc='mean').swaplevel(0,1,axis=1)
fig, axes = plt.subplots(3,4,figsize=(21,16))
items = [i[0] for i in cate.columns]
color = sns.color_palette("hls", 11)
v=0
for h in [0,1,2]:
    for c in [0,1,2,3]:
        #if h==2 and c==3:
            #break
        sns.pointplot(data = cate[items[v]],x=cate[items[v]].index,y='판매량',ax=axes[(h,c)],color=color[v])
        axes[(h,c)].set(ylabel='판매량',title=items[v])
        v+=1

fig.show()
```
